```python
from deep_translator import GoogleTranslator
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import threading
import time

class TempoEccessivoError(Exception):
    pass

def controlla_tempo(limite, driver):
    inizio = time.time()
    while True:
        tempo_trascorso = time.time() - inizio
        if tempo_trascorso > limite:
            driver.quit()
            
            raise TempoEccessivoError(f"Il tempo trascorso {tempo_trascorso} è superiore al limite {limite}")

x = []

def translate_text(text, languages):
    translations = {}
    for lang in languages:
        translated = GoogleTranslator(source='auto', target=lang).translate(text)
        translations[lang] = translated
    return translations

# Example usage
query = "specie di cani"
languages_to_translate = ['it', 'fr']  # Italian, French, German, Spanish
translations = translate_text(query, languages_to_translate)
for language, translation in translations.items():
    print(f"{language.title()}: {translation}")

    # Initialize the WebDriver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)
    driver.get(f"https://gibiru.com/results.html?q={translation}&rt=&cx=partner-pub-5956360965567042%3A8627692578&cof=FORID%3A11&ie=UTF-8")

    # Number of pages to visit
    numero_pagine = 2
    links = []

    try:
        for page_number in range(1, numero_pagine + 1):
            # Wait for the search results container to load
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "gsc-control-cse"))
            )

            # Find all <a> elements within the specified container
            search_results = driver.find_element(By.CLASS_NAME, "gsc-control-cse")
            a_tags = search_results.find_elements(By.TAG_NAME, 'a')

            # Extract href attributes from <a> tags and add them to the links list
            for tag in a_tags:
                href = tag.get_attribute('href')
                if href:
                    links.append(href)

            # Find and click the next page button
            next_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, f'div.gsc-cursor-page[aria-label="Page {page_number + 1}"]'))
            )
            next_button.click()

            # Wait a bit to allow the page to load
            time.sleep(2)

    finally:
        # Close the browser after collecting the links
        driver.quit()

    # Print all collected links
    for link in links[1:]:
        print(link)

        # Replace 'URL_DA_ESTRARRE' with the actual URL of the page you want to extract text from
        url = link

        # Perform a GET request to the web page
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            # Create a BeautifulSoup object to analyze the HTML content
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract all text from the page using the .get_text() method
            text = soup.get_text(separator=' ', strip=True)

            # Print the extracted text
            print(text)

            # Initialize the WebDriver
            driver = webdriver.Chrome(service=webdriver.chrome.service.Service(ChromeDriverManager().install()))
            driver.get('https://perplexity.ai')

            try:
                # Avvia il thread
                thread = threading.Thread(target=controlla_tempo, args=(60,driver))
                thread.start()

                # Ensure the page is fully loaded
                WebDriverWait(driver, 10).until(lambda d: d.execute_script('return document.readyState') == 'complete')
                print("Page loaded.")
                thread2 = threading.Thread(target=controlla_tempo, args=(60,driver))
                thread2.start()
                # Locate and interact with the search box
                search_box = WebDriverWait(driver, 60).until(
                    EC.visibility_of_element_located((By.CSS_SELECTOR, 'textarea[placeholder="Ask anything..."]'))
                )
                print("Search box is visible.")
                thread3 = threading.Thread(target=controlla_tempo, args=(60,driver))
                thread3.start()
                # Enter the search query
                search_query = f'il tuo compito è quello di estrarre dal seguente testo "{query}", dai quindi in output solo quanto richiesto separato da una virgola se nel testo ti viene richiesto è presente quella informazione più di una volta, testo:'+text
                search_box.send_keys(search_query)
                search_box.send_keys(Keys.RETURN)

                # Wait and retrieve the results
                results_locator = 'div[dir="auto"]'
                results = WebDriverWait(driver, 30).until(
                    EC.visibility_of_element_located((By.CSS_SELECTOR, results_locator))
                )
                print("Results are visible.")
                x.append(str(results.text))

            except Exception as e:
                print("An error occurred")

            finally:
                # Close the driver regardless of the result
                print("ciao")
                driver.quit()
                print("ciaociao")
        else:
            print(f'Error during the request: {response.status_code}')
```
